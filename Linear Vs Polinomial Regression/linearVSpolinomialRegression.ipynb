```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generating linear data
x_linear = np.linspace(0, 10, 50)
y_linear = 2 * x_linear + 5 + np.random.randn(50) # adding some noice
```

```python
# generating non-linear data
x_nonlinear = np.linspace(0, 10, 50)
y_nonlinear = 2 * np.sin(x_nonlinear) + np.random.randn(50) # adding some noice
```

```python
# linear scatter diagram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear Relationship')
plt.scatter(x_linear, 2 * x_linear + 5, color='red', label='True Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()

# non-linear scatter diagram
plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color='blue', label='Data with Non-Linear Relationship')
plt.scatter(x_linear, 2 * np.sin(x_nonlinear), color='red', label='True Non-Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-Linear Relationship')
plt.legend()
```

```python
# correlation cofficients
correlation_linear = np.corrcoef(x_linear, y_linear) [0, 1]
correlation_nonlinear = np.corrcoef(x_nonlinear, y_nonlinear) [0, 1]
```

```python
print('Correlation cofficient for linear relationship : ', correlation_linear)
print('Correlation cofficient for non-linear relationship : ', correlation_nonlinear)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg1 = LinearRegression()
```

```python
reg1.fit(x_linear.reshape(-1, 1), y_linear) # x = x_linear.reshape(-1, 1)
```

```python
reg1.score(x_linear.reshape(-1, 1), y_linear)
```

```python
# show best fit line
plt.plot(x_linear, reg1.predict(x_linear.reshape(-1, 1)))
```

```python
# best fit line with dataset value
plt.scatter(x_linear, y_linear, color="blue", label="Data with Linear Relation")

plt.plot(x_linear, reg1.predict(x_linear.reshape(-1, 1)), color="green", label="Best Fit Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
```
